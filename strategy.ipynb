```python
def calc_avg_points_per_year(data: xr.DataArray):
    if len(data.time) < 251:
        if data.name == 'crypto':
            return round(365.25 * 24)
        if data.name in ['cryptofutures', 'crypto_futures', 'crypto_daily', 'cryptodaily',
                         'crypto_daily_long', 'crypto_daily_long_short']:
            return 365
        if data.name in ['stocks', 'stocks_long', 'futures', 'stocks_nasdaq100', 'stocks_s&p500']:
            return 251
    t = np.sort(data.coords[ds.TIME].values)
    tp = np.roll(t, 1)
    dh = (t[1:] - tp[1:]).mean().item() / (10 ** 9) / 60 / 60  # avg diff in hours
    return round(365.25 * 24 / dh)

def get_env(key, def_val, silent=False):
    if key in os.environ:
        return os.environ[key]
    else:
        if not silent:
            log_err("NOTICE: The environment variable " + key + " was not specified. The default value is '" + def_val + "'")
        return def_val

def get_default_slippage(data):
    if data.name == 'stocks_s&p500':
        return float(get_env('SL_STOCKS_S&P500', '0.05', True))
    if data.name == 'stocks_nasdaq100':
        return float(get_env('SL_STOCKS_NASDAQ100', '0.05', True))
    if data.name == 'stocks':
        return float(get_env('SL_STOCKS', '0.05', True))
    if data.name == 'futures':
        return float(get_env('SL_FUTURES', '0.04', True))
    if data.name == 'cryptofutures' or data.name == 'crypto_futures':
        return float(get_env('SL_CRYPTOFUTURES', '0.04', True))
    if data.name == 'cryptodaily' or data.name == 'crypto_daily' or data.name == 'crypto_daily_long' or data.name == 'crypto_daily_long_short':
        return float(get_env('SL_CRYPTODAILY', '0.04', True))
    if data.name == 'crypto':
        return float(get_env('SL_CRYPTO', '0.05', True))
    return 0.05    

class Fields:
    OPEN = "open"
    LOW = "low"
    HIGH = "high"
    CLOSE = "close"
    VOL = "vol"
    DIVS = "divs"  # only for stocks
    SPLIT = "split"  # only for stocks
    SPLIT_CUMPROD = "split_cumprod"  # only for stocks
    IS_LIQUID = "is_liquid"  # only for stocks
    OPEN_INTEREST = "oi"  # only for futures
    ROLL = "roll"  # only for futures
    
f = Fields

class Dimensions:
    TIME = 'time'
    FIELD = 'field'
    ASSET = 'asset'

ds = Dimensions

def calc_points_per_day(days_per_year):
    if days_per_year < 400:
        return 1
    else:
        return 24
    
def calc_slippage(data, period_days=14, fract=None, points_per_year=None):
    """
    :param data: xarray with historical data
    :param period_days: period for atr
    :param fract: slippage factor
    :return: xarray with slippage
    """
    if fract is None:
        fract = get_default_slippage(data)

    if points_per_year is None:
        points_per_year = calc_avg_points_per_year(data)

    time_series = np.sort(data.coords[ds.TIME])
    data = data.transpose(ds.FIELD, ds.TIME, ds.ASSET).loc[[f.CLOSE, f.HIGH, f.LOW], time_series, :]

    points_per_day = calc_points_per_day(points_per_year)
    daily_period = min(points_per_day, len(data.time))

    cl = data.loc[f.CLOSE].shift({ds.TIME: daily_period})
    hi = data.loc[f.HIGH].rolling({ds.TIME: daily_period}).max()
    lo = data.loc[f.LOW].rolling({ds.TIME: daily_period}).min()
    d1 = hi - lo
    d2 = abs(hi - cl)
    d3 = abs(cl - lo)
    dd = xr.concat([d1, d2, d3], dim='d').max(dim='d', skipna=False)

    atr_period = min(len(dd.time), period_days * points_per_day)

    dd = dd.rolling({ds.TIME: atr_period}, min_periods=atr_period).mean(skipna=False).ffill(ds.TIME)
    return dd * fract
```

```python
# def split_data(N, B):
#     # Calculate sizes
#     train_pct=1
#     N_train = int(train_pct * N)
#     N_valid = int((1-train_pct) * N)
#     N_test = N - N_train - N_valid
    
#     # Generate main ranges
#     train_range = range(0, N_train)
#     valid_range = range(N_train, N_train + N_valid)
#     test_range = range(N_train + N_valid, N)
    
#     # Calculate the size of each batch
#     N_batch = math.ceil(N_train / B)
    
#     # Generate batch ranges
#     batch_ranges = [range(i, min(i + N_batch, N_train)) for i in range(0, N_train, N_batch)]
    
#     return train_range, valid_range, test_range, batch_ranges

def average_price_over_time(prices, quantities):
    # Compute cumulative weighted sum
    weighted_sum = tf.math.cumsum(prices * quantities, axis=0)  # Shape (10, 3)
    # Compute cumulative quantity
    total_quantity = tf.math.cumsum(quantities, axis=0)  # Shape (10, 3)
    avg_price = tf.math.divide_no_nan(weighted_sum, total_quantity)
    return avg_price

def total_quantity(quantities):
    # Compute cumulative quantity
    total_quantity = tf.math.cumsum(quantities, axis=0)  # Shape (10, 3)
    return total_quantity    

def avg_holding_sign_change(quantities):
    # Compute the sign of the total position (long/short)
    sign_total_quantity = tf.sign(quantities)  # -1 for short, 1 for long, 0 if exactly zero
    # Shifted sign to detect changes (prepend a row of zeros for correct diff computation)
    sign_shifted = tf.concat([tf.zeros((1, tf.shape(quantities)[1]), dtype=tf.float32), sign_total_quantity[:-1]], axis=0)
    # Detect sign changes (where the sign of total_quantity changes)
    sign_change = tf.cast(sign_total_quantity != sign_shifted, tf.float32)
    #res = tf.math.cumsum(sign_change, axis=0)/tf.range(1, tf.shape(sign_change)[0] + 1, dtype=tf.float32)[:, tf.newaxis] # FS COMM: calcu denominator inadvance outside
    res=tf.math.divide_no_nan( tf.math.cumsum(sign_change, axis=0),tf.range(1, tf.shape(sign_change)[0] + 1, dtype=tf.float32))[:, tf.newaxis]
    return res

def position_size_volatility(quantities):
    # Compute cumulative mean
    cum_mean = tf.cumsum(quantities, axis=0) / tf.range(1, tf.shape(quantities)[0] + 1, dtype=tf.float32)[:, tf.newaxis]
    # Compute squared deviations from the mean
    squared_deviation = tf.square(quantities - cum_mean)
    # Compute cumulative variance
    #cum_variance = tf.cumsum(squared_deviation, axis=0) / tf.range(1, tf.shape(quantities)[0] + 1, dtype=tf.float32)[:, tf.newaxis]
    cum_variance = tf.math.divide_no_nan( tf.cumsum(squared_deviation, axis=0), tf.range(1, tf.shape(quantities)[0] + 1, dtype=tf.float32))[:, tf.newaxis]
    # Compute rolling volatility (std deviation)
    volatility = tf.sqrt(cum_variance)
    return volatility

def portfolio_turnover_rate(quantities):
    # Compute absolute changes in position size
    delta_quantities = tf.abs(tf.concat([tf.zeros((1, tf.shape(quantities)[1]), dtype=tf.float32), quantities[1:] - quantities[:-1]], axis=0))
    # Compute cumulative absolute position size
    total_position = tf.math.cumsum(tf.abs(quantities), axis=0)
    # Compute cumulative turnover
    cumulative_turnover = tf.math.cumsum(delta_quantities, axis=0)
    turnover_rate = tf.math.divide_no_nan(cumulative_turnover, total_position)
    return turnover_rate

def holding_duration(quantities):
    """
    Computes how many consecutive time steps each asset has been held without being fully sold.
    Args:
        quantities (tf.Tensor): A 2D tensor of shape [time, assets] with asset quantities.
    Returns:
        tf.Tensor: A 2D tensor of the same shape representing holding duration per asset.
    """
    quantities = tf.cast(quantities > 0, tf.float32)  # Binary: 1 if held, 0 if not
    def scan1_fn(prev, current):
        # If asset is held (current == 1), increment streak; else reset to 0
        return tf.where(current > 0, prev + 1, tf.zeros_like(current))
    # Use tf.scan to accumulate holding streaks over time
    durations = tf.scan(scan1_fn, quantities, initializer=tf.zeros_like(quantities[0]))
    return durations

def position_change_magnitude(quantities):
    """
    Computes the absolute change in quantity (buy/sell size) for each asset at each time step.
    Args:
        quantities (tf.Tensor): A [time, assets] tensor of asset quantities.
    Returns:
        tf.Tensor: A [time, assets] tensor of absolute changes. First row is zero.
    """
    # Pad with zeros at the beginning so the shape stays [time, assets]
    prev = tf.concat([tf.cast(tf.zeros_like(quantities[:1]),tf.float32), quantities[:-1]], axis=0)
    # Absolute change from previous step
    change = tf.abs(quantities - prev)
    return change

def weighted_exposure(prices, quantities):
    pv = tf.reduce_sum(prices * quantities, axis=1, keepdims=True)
    exposure = tf.math.divide_no_nan(prices * quantities, pv)
    return exposure   # already [time, assets]


def herfindahl_index(quantities):
    total = tf.reduce_sum(tf.abs(quantities), axis=1, keepdims=True)
    weights = tf.math.divide_no_nan(tf.abs(quantities), total)
    hi = tf.reduce_sum(tf.square(weights), axis=1, keepdims=True)
    hi = tf.tile(hi, [1, tf.shape(quantities)[1]])  # broadcast to assets
    return hi

def average_holding_size(quantities):
    cum_sum = tf.cumsum(tf.abs(quantities), axis=0)
    counts = tf.range(1, tf.shape(quantities)[0]+1, dtype=tf.float32)[:, tf.newaxis]
    return tf.math.divide_no_nan(cum_sum, counts)

def position_change_frequency(quantities):
    delta = tf.abs(quantities[1:] - quantities[:-1])
    change = tf.cast(delta > 0, tf.float32)
    freq = tf.cumsum(change, axis=0) / tf.cast(tf.range(1, tf.shape(quantities)[0]), tf.float32)[:, tf.newaxis]
    # Pad first row with zeros
    freq = tf.concat([tf.zeros((1, tf.shape(quantities)[1])), freq], axis=0)
    return freq

def short_long_ratio(quantities):
    shorts = tf.reduce_sum(tf.cast(quantities < 0, tf.float32), axis=1, keepdims=True)
    longs  = tf.reduce_sum(tf.cast(quantities > 0, tf.float32), axis=1, keepdims=True)
    ratio  = tf.math.divide_no_nan(shorts, longs)
    ratio = tf.tile(ratio, [1, tf.shape(quantities)[1]])
    return ratio

def weighted_holding_duration(quantities):
    """
    Weighted average holding duration, broadcast per asset.
    Output shape: [time, assets]
    """
    # [time, assets]
    durations = holding_duration(quantities)
    weights = tf.abs(quantities)
    # [time, 1]
    total_weight = tf.reduce_sum(weights, axis=1, keepdims=True)
    weighted_sum = tf.reduce_sum(durations * weights, axis=1, keepdims=True)
    # [time, 1]
    whd = tf.math.divide_no_nan(weighted_sum, total_weight)
    # 🔑 CRITICAL LINE: broadcast to assets
    whd = tf.tile(whd, [1, tf.shape(quantities)[1]])
    return whd


# Define a single model that handles both passes properly
class FeedbackNN(tf.keras.Model):
    def __init__(self, input_dim):
        super(FeedbackNN, self).__init__()

        self.feature_extractor = tf.keras.Sequential([
            layers.Input(shape=(numTechInd,)), 
            layers.LayerNormalization(),
            tf.keras.layers.Dense(24, activation='linear',kernel_initializer='glorot_uniform'),             
            tf.keras.layers.LeakyReLU(negative_slope=negative_slope),  
            tf.keras.layers.Dense(8, activation='linear',kernel_initializer='glorot_uniform'),      
            tf.keras.layers.LeakyReLU(negative_slope=negative_slope),    
            tf.keras.layers.Dense(12, activation='linear',kernel_initializer='glorot_uniform'),      
            tf.keras.layers.LeakyReLU(negative_slope=negative_slope),        
            tf.keras.layers.Dense(1, activation='linear',kernel_initializer='glorot_uniform')  
        ])
        
        self.final_predictor = tf.keras.Sequential([
            layers.Input(shape=(10,)), ## FS PARAM
            tf.keras.layers.Dense(8, activation='linear',kernel_initializer='glorot_uniform'),             
            tf.keras.layers.LeakyReLU(negative_slope=negative_slope),  
            tf.keras.layers.Dense(4, activation='linear',kernel_initializer='glorot_uniform'),      
            tf.keras.layers.LeakyReLU(negative_slope=negative_slope),                                                
            tf.keras.layers.Dense(1, activation='linear',kernel_initializer='glorot_uniform')  
        ])

    def call(self, inputs, training=False):
        # First pass: Generate derived features (shape: batch_size x 10)
        rwf_intleav_tf = self.feature_extractor(inputs, training=training)
        NN_wf_tf=tf.reshape(rwf_intleav_tf, (num_time_batch, num_assets))

        ## FS TEST - remove norm
        ### Normalize as I do in becnhmarking Test_weights=(Test_weights/abs(Test_weights).sum(ds.ASSET)).fillna(0)
        NN_norm_sum_abs = tf.reduce_sum(tf.abs(NN_wf_tf), axis=1, keepdims=True)
        NN_norm_weights = tf.math.divide_no_nan(NN_wf_tf, NN_norm_sum_abs)
        ## Normalize and cap as does in Quantiacs 
        NN_sum_abs = tf.reduce_sum(tf.abs(NN_norm_weights),axis=1)
        NN_s = tf.expand_dims(tf.maximum(NN_sum_abs, 1.0), axis=-1)
        NN_raw_target_weights = tf.math.divide_no_nan(NN_norm_weights,NN_s)
        NN_W_tf = tf.roll(NN_raw_target_weights, shift=1, axis=0)

        NN_W_tf = tf.tensor_scatter_nd_update(
            NN_W_tf,
            indices=[[0]],                    
            updates=tf.zeros((1, num_assets), dtype=tf.float32)  
        )       
        
        scan_inputs_NN = (time_indices,
                       open_loop_tf,
                       open_loop_tm1_tf,
                       close_loop_tf,
                       close_loop_tm1_tf,
                       divs_loop_tf,
                       UNLOCKED_loop_tf,
                       slippage_loop_tf,
                       NN_W_tf
                      ) 
        # #print (f"scan_inputs: {scan_inputs}")
        #print ("strat tf.scan in NN")        
        results = tf.scan(step_NN_fn, scan_inputs_NN, initializer=initial_state_NN) # Run `tf.scan` to perform sequential updates efficiently
        #print (f"NN results:{results}")
        dN = results["dN"] # Extract daily quantities    
        
        avg_price_paid=tf.reshape(average_price_over_time(open_loop_tf, dN), (num_time_batch*num_assets, 1))
        pre_open_tf = tf.roll(open_loop_tf, shift=1, axis=0)
        pre_open_tf = tf.reshape(tf.tensor_scatter_nd_update(
            pre_open_tf,
            indices=[[0]],                   
            updates=tf.zeros((1, num_assets), dtype=tf.float32)),(num_time_batch*num_assets, 1))
        total_quantity_kpi=tf.reshape(total_quantity(dN), (num_time_batch*num_assets, 1))  
        turnover_rate_kpi=tf.reshape(portfolio_turnover_rate(dN), (num_time_batch*num_assets, 1))
        holding_duration_kpi=tf.reshape(holding_duration(dN), (num_time_batch*num_assets, 1))       
        position_change_magnitude_kpi=tf.reshape(position_change_magnitude(dN), (num_time_batch*num_assets, 1))
        weighted_exposure_kpi = tf.reshape(weighted_exposure(open_loop_tf,dN), (num_time_batch*num_assets, 1))
        # herfindahl_index_kpi = tf.reshape(herfindahl_index(dN), (num_time_batch*num_assets, 1))
        average_holding_size_kpi = tf.reshape(average_holding_size(dN), (num_time_batch*num_assets, 1))
        position_change_frequency_kpi = tf.reshape(position_change_frequency(dN), (num_time_batch*num_assets, 1))
        short_long_ratio_kpi = tf.reshape(short_long_ratio(dN), (num_time_batch*num_assets, 1))
        
        combined_input = tf.concat([avg_price_paid,
                                    pre_open_tf,
                                    total_quantity_kpi,
                                    turnover_rate_kpi,
                                    holding_duration_kpi,
                                    position_change_magnitude_kpi,
                                    weighted_exposure_kpi,
                                    # herfindahl_index_kpi,
                                    average_holding_size_kpi,
                                    position_change_frequency_kpi,
                                    short_long_ratio_kpi
                                   ], axis=1)
        # Second pass: Generate final output
        raw_weigths = self.final_predictor(combined_input, training=training)

        return raw_weigths

modelC1 = FeedbackNN(numTechInd)

# test to remove the loop with tensor  op
# Define step function to be used in `tf.scan`
## TEST SCAN FIRST - to benchmark - without using teh gradient train loop
def step_fn(prev_state, inputs):
    t, open_t, open_tm1, close_t, close_tm1, divs_t, UNLOCKED_t, slippage_t, weights_t  = inputs
    #print (f"time t : {t}")
    #print (f"open_t: {open_t}")
    prev_shares_count = prev_state["shares_count"]
    #tf.debugging.check_numerics(prev_shares_count, "prev_shares_count contains NaN or Inf")
    #print (f"prev_shares_count: {prev_shares_count}")
    prev_equity_after_buy = prev_state["equity_after_buy"]

    ### equity_before_buy ###############################
    pnl = tf.reduce_sum((open_t - open_tm1 + divs_t) * prev_shares_count)
    equity_before_buy = tf.where(
        t > 0,
        prev_equity_after_buy + pnl,
        tf.ones_like(prev_equity_after_buy)
    )  
    #print (f"equity_before_buy: {equity_before_buy}")
    #tf.debugging.check_numerics(equity_before_buy, "equity_before_buy contains NaN or Inf")         

    ## target_weights ##############################    
    target_weights = tf.math.divide_no_nan(weights_t * open_t, close_tm1)    
    sum_abs_t = tf.reduce_sum(tf.abs(target_weights))
    st = tf.maximum(sum_abs_t, 1.0)
    target_weights = tf.math.divide_no_nan(target_weights,st)
    # if t<210:
    #     print (f"t: {t}, target_weights: {target_weights}")

    ## shares_count ##############
    w_free_cash = st - sum_abs_t
    w_unlocked = tf.reduce_sum(tf.abs(target_weights*UNLOCKED_t))
    w_operable = w_unlocked + w_free_cash
    equity_operable_before_buy = equity_before_buy - tf.reduce_sum(open_t*(1-UNLOCKED_t)*tf.abs(prev_shares_count))
    #print (w_free_cash.shape,w_unlocked.shape,w_operable.shape, UNLOCKED_t.shape,target_weights.shape )
    shares_count = tf.where(w_operable < EPS,
                            prev_shares_count,
                            tf.math.divide_no_nan(equity_operable_before_buy*target_weights*UNLOCKED_t,
                                                  w_operable*UNLOCKED_t*open_t))   
    
    ## dN ##############
    dN = tf.where(
            t > 0,
            shares_count - prev_shares_count*UNLOCKED_t,
            shares_count
        )                     
    #print (f"dN: {dN}")   
    # Compute `step_slippage`
    ## equity_tonight ##############
    step_slippage = tf.reduce_sum(slippage_t*UNLOCKED_t* tf.abs(dN))
    equity_after_buy = tf.where(w_operable < EPS,
                               equity_before_buy,
                               equity_before_buy - step_slippage)    
    equity_tonight = equity_after_buy + tf.reduce_sum((close_t - open_t) * shares_count)
    #print (f"equity_tonight: {equity_tonight}")   

    return {
        "equity_operable_before_buy":equity_operable_before_buy,
        "shares_count": shares_count,
        "equity_before_buy": equity_before_buy,
        "equity_after_buy": equity_after_buy,
        "equity_tonight": equity_tonight,
        "dN":dN,
        "target_weights": target_weights
    }

## TEST SCAN FIRST - to benchmark - without using teh gradient train loop
def step_NN_fn(prev_state, inputs):
    t, open_t, open_tm1, close_t, close_tm1, divs_t, UNLOCKED_t, slippage_t, weights_t  = inputs
    prev_shares_count = prev_state["shares_count"]
    prev_equity_after_buy = prev_state["equity_after_buy"]
 
    ### equity_before_buy ###############################
    pnl = tf.reduce_sum((open_t - open_tm1 + divs_t) * prev_shares_count)
    equity_before_buy = tf.where(
        t > 0,
        prev_equity_after_buy + pnl,
        tf.ones_like(prev_equity_after_buy)
    )  
    #print (f"equity_before_buy: {equity_before_buy}")
    #tf.debugging.check_numerics(equity_before_buy, "equity_before_buy contains NaN or Inf")     

    ## target_weights ##############################    
    target_weights = tf.math.divide_no_nan(weights_t * open_t, close_tm1)    
    sum_abs_t = tf.reduce_sum(tf.abs(target_weights))
    st = tf.maximum(sum_abs_t, 1.0)
    target_weights = tf.math.divide_no_nan(target_weights,st)
    # if t<210:
    #     print (f"t: {t}, target_weights: {target_weights}")
    
    ## shares_count ##############
    w_free_cash = st - sum_abs_t
    w_unlocked = tf.reduce_sum(tf.abs(target_weights*UNLOCKED_t))
    w_operable = w_unlocked + w_free_cash
    equity_operable_before_buy = equity_before_buy - tf.reduce_sum(open_t*(1-UNLOCKED_t)*tf.abs(prev_shares_count))
    #print (w_free_cash.shape,w_unlocked.shape,w_operable.shape, UNLOCKED_t.shape,target_weights.shape )
    shares_count = tf.where(w_operable < EPS,
                            prev_shares_count,
                            tf.math.divide_no_nan(equity_operable_before_buy*target_weights*UNLOCKED_t,
                                                  w_operable*UNLOCKED_t*open_t))   

    ## dN ##############
    dN = tf.where(
            t > 0,
            shares_count - prev_shares_count*UNLOCKED_t,
            shares_count
        )                     
    ## equity_after_buy ##############
    step_slippage = tf.reduce_sum(slippage_t*UNLOCKED_t* tf.abs(dN))
    equity_after_buy = tf.where(w_operable < EPS,
                               equity_before_buy,
                               equity_before_buy - step_slippage)  
    
    return {
        "equity_operable_before_buy":equity_operable_before_buy,
        "shares_count": shares_count,
        "equity_before_buy": equity_before_buy,
        "equity_after_buy": equity_after_buy,
        "dN":dN,
        "target_weights": target_weights
    }    

#train_range, valid_range, test_range, batch_ranges = split_data((num_time//batch_len)*batch_len, num_time//batch_len)
batch_ranges = [range(0, num_time)]
# Assuming modelC1 is your TensorFlow/Keras model
num_params = sum([tf.size(var).numpy() for var in modelC1.trainable_variables])
print("Total number of trainable parameters:", num_params)
```

```python
from qnt.output import normalize
from qnt.output import normalize as output_normalize
Q_equity_tonight_lst=[]
Q_equity_before_buy_lst= []
Q_shares_count_lst= []
Q_equity_operable_before_buy_lst=[]
Q_equity_after_buy_lst=[]
Q_step_slippage_lst=[]
Q_dN_lst=[]
Q_target_weights, Q_arrange_target_weights, Q_portfolio_history = None, None, None  # define global first

def arrange_data(data, target_weights, per_asset, *additional_series):
    """
    arranges data for proper calculations
    :param per_asset:
    :param data:
    :param target_weights:
    :param additional_series:
    :return:
    """
    data = data.dropna(ds.ASSET, how='all').dropna(ds.TIME, how='all')
    target_weights, data = xr.align(target_weights, data, join='right')
    ra1 = []
    for a in additional_series:
        a, _ = xr.align(a, data, join='right')
        ra1.append(a)

    time_series = data.coords[ds.TIME]
    time_series = np.sort(time_series)

    assets = data.coords[ds.ASSET].values
    assets = np.sort(assets)

    adjusted_data = data.transpose(ds.FIELD, ds.TIME, ds.ASSET)
    adjusted_data = adjusted_data.loc[:, time_series, assets]

    target_weights = target_weights.transpose(ds.TIME, ds.ASSET)
    target_weights = target_weights.loc[time_series, assets]
    target_weights = normalize(target_weights, per_asset)

    ra2 = []
    for a in list(ra1):
        a = a.sel(time=time_series)
        if ds.ASSET in a.dims:
            a = a.sel(asset=assets)
        ra2.append(a)

    target_weights = target_weights.drop_vars(ds.FIELD, errors='ignore')
    #print (f"target_weights:{target_weights}")
    return (adjusted_data, target_weights, *ra2)

def calc_relative_return(data, portfolio_history,
                         slippage_factor=None, roll_slippage_factor=None,
                         per_asset=False, points_per_year=None):
    global Q_target_weights, Q_portfolio_history, Q_arrange_target_weights
    if slippage_factor is None:
        slippage_factor = get_default_slippage(data)

    if roll_slippage_factor is None:
        roll_slippage_factor = get_default_slippage(data)

    Q_portfolio_history = portfolio_history         
    target_weights = portfolio_history.shift(**{ds.TIME: 1})  # shift
    target_weights[{ds.TIME: 0}] = 0
    min_time = target_weights.coords[ds.TIME].min()

    slippage = calc_slippage(data, 14, slippage_factor, points_per_year=points_per_year)
    roll_slippage = calc_slippage(data, 14, roll_slippage_factor, points_per_year=points_per_year)

    data, target_weights, slippage, roll_slippage = arrange_data(data, target_weights, per_asset, slippage,
                                                                 roll_slippage)
    Q_arrange_target_weights = target_weights 
    # adjust weights according to price changes close->open : W_open = W_close_prev * OPEN / CLOSE_prev
    prev_close = data.loc[f.CLOSE].shift(**{ds.TIME: 1}).ffill(ds.TIME)
    target_weights = target_weights * data.loc[f.OPEN] / prev_close
    target_weights = output_normalize(target_weights, per_asset)
    #print (f"target_weights: {target_weights}")
    Q_target_weights = target_weights 

    W = target_weights
    D = data

    OPEN = D.loc[f.OPEN].ffill(ds.TIME).fillna(0)
    CLOSE = D.loc[f.CLOSE].ffill(ds.TIME).fillna(0)
    DIVS = D.loc[f.DIVS].fillna(0) if f.DIVS in D.coords[ds.FIELD] else xr.full_like(D.loc[f.CLOSE], 0)
    ROLL = D.loc[f.ROLL].fillna(0) if f.ROLL in D.coords[ds.FIELD] else None
    ROLL_SLIPPAGE = roll_slippage.where(ROLL != 0).fillna(0) if ROLL is not None else None

    # boolean matrix when assets available for trading
    UNLOCKED = np.logical_and(np.isfinite(D.loc[f.OPEN].values), np.isfinite(D.loc[f.CLOSE].values))
    #UNLOCKED = np.logical_and(np.isfinite(W.values), UNLOCKED)
    UNLOCKED = np.logical_and(np.isfinite(slippage.values), UNLOCKED)
    UNLOCKED = np.logical_and(OPEN > EPS, UNLOCKED)

    #print ('calc_relative_return')
    #print ('UNLOCKED')
    #print (UNLOCKED)

    if per_asset:
        RR = W.copy(True)
        RR[:] = calc_relative_return_np_per_asset(W.values, UNLOCKED.values, OPEN.values, CLOSE.values, slippage.values,
                                                  DIVS.values,
                                                  ROLL.values if ROLL is not None else None,
                                                  ROLL_SLIPPAGE.values if ROLL_SLIPPAGE is not None else None
                                                  )
        return RR.loc[min_time:]
    else:
        RR = xr.DataArray(
            np.full([len(W.coords[ds.TIME])], np.nan, np.double),
            dims=[ds.TIME],
            coords={ds.TIME: W.coords[ds.TIME]}
        )
        res = calc_relative_return_np(W.values, UNLOCKED.values, OPEN.values, CLOSE.values, slippage.values,
                                      DIVS.values,
                                      ROLL.values if ROLL is not None else None,
                                      ROLL_SLIPPAGE.values if ROLL_SLIPPAGE is not None else None
                                      )
        RR[:] = res
        return RR.loc[min_time:]
        
def calc_relative_return_np(WEIGHT, UNLOCKED, OPEN, CLOSE, SLIPPAGE, DIVS, ROLL, ROLL_SLIPPAGE):
    global Q_equity_tonight_lst, Q_equity_before_buy_lst, Q_shares_count_lst, Q_equity_operable_before_buy_lst,Q_equity_after_buy_lst,Q_step_slippage_lst,Q_dN_lst
    shares_count = np.zeros(WEIGHT.shape)
    equity_before_buy = np.zeros(WEIGHT.shape[0])
    equity_operable_before_buy = np.zeros(WEIGHT.shape[0])
    equity_after_buy = np.zeros(WEIGHT.shape[0])
    equity_tonight = np.zeros(WEIGHT.shape[0])
    #print (UNLOCKED.shape)
    t=0
    for day in range(WEIGHT.shape[0]):
        tradeable_assets = UNLOCKED[day]
        non_tradeable_assets = np.logical_not(tradeable_assets)

        if day == 0:
            equity_before_buy[0] = 1
            shares_count[0] = 0
        else:
            shares_count[day] = shares_count[day - 1]
            equity_before_buy[day] = equity_after_buy[day - 1] + np.nansum(
                (OPEN[day] - OPEN[day - 1] + DIVS[day]) * shares_count[day])

        w_sum = np.nansum(np.abs(WEIGHT[day]))
        w_free_cash = max(1, w_sum) - w_sum
        w_unlocked = np.nansum(np.abs(WEIGHT[day][tradeable_assets]))
        #print (f"tradeable_assets: {tradeable_assets}")
        w_operable = w_unlocked + w_free_cash

        equity_operable_before_buy[day] = equity_before_buy[day] - np.nansum(
            OPEN[day][non_tradeable_assets] * np.abs(shares_count[day][non_tradeable_assets]))

        if w_operable < EPS:
            equity_after_buy[day] = equity_before_buy[day]
        else:
            shares_count[day][tradeable_assets] = equity_operable_before_buy[day] * WEIGHT[day][tradeable_assets] / (
                    w_operable * OPEN[day][tradeable_assets])
            dN = shares_count[day][tradeable_assets]
            if day > 0:
                dN = dN - shares_count[day - 1][tradeable_assets]
            step_slippage = np.nansum(SLIPPAGE[day][tradeable_assets] * np.abs(dN))
            equity_after_buy[day] = equity_before_buy[day] - step_slippage

        if ROLL is not None and day > 0:
            partial_shares = np.where(np.sign(shares_count[day]) == np.sign(shares_count[day - 1]),
                                      np.minimum(np.abs(shares_count[day]), np.abs(shares_count[day - 1])), 0)
            roll_costs = np.sign(shares_count[day]) * partial_shares * ROLL[day] + partial_shares * ROLL_SLIPPAGE[day]
            equity_after_buy[day] -= np.nansum(roll_costs)

        equity_tonight[day] = equity_after_buy[day] + np.nansum((CLOSE[day] - OPEN[day]) * shares_count[day])
        #Q_dN_lst.append(dN.tolist())
        #print (f"WEIGHT[day]")
        # if t in [15]:
        #     #print (f"step_slippage:{step_slippage}")
        #     print (f"t:{t} , {WEIGHT[day]}")
        # Set shares count to 0 for assets without a price. Reasons: company could've ceased to exist, been acquired, or there's a provider error
        shares_count[day][non_tradeable_assets] = 0
        Q_dN_lst.append(dN.tolist())
        t+=1
    Q_equity_tonight_lst = equity_tonight
    Q_equity_before_buy_lst = equity_before_buy
    Q_shares_count_lst = shares_count.tolist()
    Q_equity_operable_before_buy_lst = equity_operable_before_buy.tolist()
    Q_equity_after_buy_lst = equity_after_buy.tolist()
    Q_step_slippage_lst=step_slippage.tolist()
    
    E = equity_tonight
    Ep = np.roll(E, 1)
    Ep[0] = 1
    RR = E / Ep - 1
    RR = np.where(np.isfinite(RR), RR, 0)
    return RR

def calc_slippage(data, period_days=14, fract=None, points_per_year=None):
    """
    :param data: xarray with historical data
    :param period_days: period for atr
    :param fract: slippage factor
    :return: xarray with slippage
    """
    if fract is None:
        fract = get_default_slippage(data)

    if points_per_year is None:
        points_per_year = calc_avg_points_per_year(data)

    time_series = np.sort(data.coords[ds.TIME])
    data = data.transpose(ds.FIELD, ds.TIME, ds.ASSET).loc[[f.CLOSE, f.HIGH, f.LOW], time_series, :]

    points_per_day = calc_points_per_day(points_per_year)
    daily_period = min(points_per_day, len(data.time))

    cl = data.loc[f.CLOSE].shift({ds.TIME: daily_period})
    hi = data.loc[f.HIGH].rolling({ds.TIME: daily_period}).max()
    lo = data.loc[f.LOW].rolling({ds.TIME: daily_period}).min()
    d1 = hi - lo
    d2 = abs(hi - cl)
    d3 = abs(cl - lo)
    dd = xr.concat([d1, d2, d3], dim='d').max(dim='d', skipna=False)

    atr_period = min(len(dd.time), period_days * points_per_day)
    #print (f"atr_period*fract: {atr_period*fract}")    

    dd = dd.rolling({ds.TIME: atr_period}, min_periods=atr_period).mean(skipna=False).ffill(ds.TIME)
    return dd * fract    

def scatterplot(x,y):
    plt.scatter(x, y)
    # plt.xlabel('')
    # plt.ylabel('')
    plt.title('')
    plt.grid(True)
    plt.show()
    # Calculate the correlation coefficient
    correlation_coefficient = np.corrcoef(x, y)[0, 1]
    return plt.show(),print("Correlation Coefficient:", correlation_coefficient)    

def calc_equity(relative_return):
    """
    :param relative_return: daily return
    :return: daily portfolio equity
    """
    return (relative_return + 1).cumprod(ds.TIME)    
```
